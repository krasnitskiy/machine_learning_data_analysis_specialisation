```python
import math
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
# Create function
def f(x):
    return math.sin(x/5)*math.exp(x/10)+5*math.exp(-x/2)
```

```python
# Graph for our function
x_list = np.arange(1, 30, 0.1)
y_list = [f(x) for x in x_list]
plt.plot(x_list, y_list)
plt.show()
```

```python
optimize.differential_evolution(f, [(1, 30)]) # global minimum calculation by differential evolution method
```

We can see that differential evolution algorithm found global minimum correctly because it intented for solve local minimum problem.
