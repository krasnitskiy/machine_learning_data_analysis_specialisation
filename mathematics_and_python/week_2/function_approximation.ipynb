```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Initial function
def f(x):
    return math.sin(x/5)*math.exp(x/10)+5*math.exp(-x/2)
```

### $$w_0 + w_1x$$

```python
# Create and solve linear equation system 
a = np.array([[1, 1], [1, 15]])
b = np.array([f(1), f(15)])
w = np.linalg.solve(a, b)
```

```python
# Result 
w
```

```python
# Graphs
x_list = np.arange(0, 16, 1)
y_list = [f(x) for x in x_list]
plt.plot(x_list, w[0] + w[1]*x_list)
plt.plot(x_list, y_list)
plt.title('Approximation')
```

### $$w_0 + w_1x + w_2x^2$$

```python
# Create and solve linear equation system 
a = np.array([[1, 1, 1], [1, 8, 64], [1, 15, 225]])
b = np.array([f(1), f(8), f(15)])
w = np.linalg.solve(a, b)
```

```python
# Result 
w
```

```python
# Graphs
x_list = np.arange(0, 16, 1)
y_list = [f(x) for x in x_list]
plt.plot(x_list, w[0] + w[1]*x_list + w[2]*(x_list**2))
plt.plot(x_list, y_list)
plt.title('Approximation')
```

### $$w_0 + w_1x + w_2x^2 + w_3x^3$$

```python
# Create and solve linear equation system 
a = np.array([[1, 1, 1, 1], [1, 4, 16, 64], [1, 10, 100, 1000], [1, 15, 225, 3375]])
b = np.array([f(1), f(4), f(10), f(15)])
w = np.linalg.solve(a, b)
```

```python
# Result 
w
```

```python
# Graphs
x_list = np.arange(0, 16, 1)
y_list = [f(x) for x in x_list]
plt.plot(x_list, w[0] + w[1]*x_list + w[2]*(x_list**2) +w[3]*(x_list**3))
plt.plot(x_list, y_list)
plt.title('Approximation')
```
